# Plot data HTI 2D

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

# Load Full data

```python
list_full = [line.strip() for line in open("Out_temperature.txt", 'r')]
```

# Load komponen x dan z

```python
#mengambil nilai x
x = []
it = 1
temp = list_full[5*it].split() #menyiapkan temporary variable
while temp[0] != 'Y-Direction':
    temp = list_full[5*it].split()
    if temp[0] != 'Y-Direction':
        x = x + temp
    it = it + 1

z = []
it = it + 1
temp = list_full[5*it+1].split()
while len(temp) > 0:
    temp = list_full[5*it+1].split()
    if len(temp) > 0:
        z = z + temp
    it = it + 1
```

# Plot titik x dan z

```python
x_arr = np.array(x,dtype='float')
z_arr = np.array(z,dtype='float')
xx,zz = np.meshgrid(x_arr,z_arr)
plt.figure(figsize=(5,2),dpi=200)
plt.scatter(xx,-zz,s=0.1)
plt.axis('equal')
plt.show()
```

# Load distribusi nilai temperatur 2D untuk n_tstep

```python
it = 0
while list_full[it] != '--- Temperature Values ---  (Deg.C)':
    it = it +1
    
it = it + 2
val = []
n = 12
len_row = 122
n_tstep = 2

mul = int(np.ceil(len(x)/11))+2
data_val = np.empty((len(z),mul*10))
data_val_tstep = np.empty((len(z),len(x),n_tstep))
for it2 in range(n_tstep):
    for it1 in range(mul):
        data_val_1 = []
        for it in range(it+1,it+len(z)+1):
            list_row = [(list_full[it][i:i+n]).strip() for i in range(2, len_row, n)]
            list_row_val = ['-999' if i=='' else i for i in list_row]
            data_val_1.append(np.array(list_row_val,dtype='float'))
            it = it+4
        data_val_1 = np.array(data_val_1)
        data_val[:,it1*10:(it1+1)*10] = data_val_1
        
    it = it+4
    data_val_2d = data_val.reshape(len(z),-1)
    data_val_2d [ data_val_2d == -999] = np.nan
    data_val_2d = data_val_2d[:,:len(x)]
    data_val_tstep[:,:,it2] = data_val_2d
    
```

# Plot kontur distribusi nilai temperatur 2D untuk n_tstep

```python
plt.figure(figsize=(5,n_tstep*2),dpi=200)
current_cmap = matplotlib.cm.get_cmap()
current_cmap.set_bad(color='white')
for it3 in range(n_tstep):
    plt.subplot(n_tstep,1,it3+1)
    plt.contourf(xx,-zz,data_val_tstep[:,:,it3],cmap='jet',vmin=0,levels=50)
    plt.axis('equal')
plt.show()
```
